### Implement the `coffee_expert_system` function

We will now write the `coffee_expert_system` function which will take five boolean parameters representing user preferences: `prefers_sweet`, `prefers_milk`, `prefers_strong`, `prefers_iced`, and `prefers_decaf`. Based on the combinations of these preferences, the function will apply a set of `if-elif-else` rules to recommend a specific coffee drink. A default recommendation will be provided if no specific rule matches the input preferences.

**Reasoning**:
Now, I will implement the `coffee_expert_system` function using `if-elif-else` statements to apply the decision rules based on the user's preferences, ensuring a default recommendation is provided if no specific rule matches.



```python
def coffee_expert_system(prefers_sweet, prefers_milk, prefers_strong, prefers_iced, prefers_decaf):
    """
    Recommends a coffee drink based on user preferences.

    Args:
        prefers_sweet (bool): True if the user prefers sweet coffee.
        prefers_milk (bool): True if the user prefers coffee with milk.
        prefers_strong (bool): True if the user prefers strong coffee.
        prefers_iced (bool): True if the user prefers iced coffee.
        prefers_decaf (bool): True if the user prefers decaffeinated coffee.

    Returns:
        str: A recommended coffee drink.
    """
    if prefers_decaf:
        if prefers_iced and prefers_milk and prefers_sweet:
            return "Iced Decaf Vanilla Latte"
        elif prefers_milk and prefers_sweet:
            return "Decaf Caramel Macchiato"
        elif prefers_strong:
            return "Decaf Americano"
        else:
            return "Decaf Coffee"

    if prefers_iced:
        if prefers_sweet and prefers_milk:
            return "Iced Mocha"
        elif prefers_milk:
            return "Iced Latte"
        elif prefers_strong:
            return "Iced Americano"
        else:
            return "Cold Brew"

    if prefers_strong:
        if prefers_sweet and prefers_milk:
            return "Espresso Con Panna"
        elif prefers_milk:
            return "Flat White"
        else:
            return "Espresso Shot"

    if prefers_milk:
        if prefers_sweet:
            return "Cappuccino"
        else:
            return "Latte"

    if prefers_sweet:
        return "Regular Coffee with Sugar"

    return "Regular Coffee"

print("The coffee_expert_system function has been defined.")
```

### Test the Expert System

Now, let's test the `coffee_expert_system` function with a few different combinations of user preferences to see if it provides the expected recommendations. This will help verify that our decision rules are correctly implemented.

**Reasoning**:

```python
print("Testing coffee_expert_system with various preferences:")

# Test case 1: Sweet, milk, no strong, no iced, no decaf
recommendation = coffee_expert_system(prefers_sweet=True, prefers_milk=True, prefers_strong=False, prefers_iced=False, prefers_decaf=False)
print(f"Sweet: True, Milk: True, Strong: False, Iced: False, Decaf: False -> Recommendation: {recommendation}")

# Test case 2: Strong, no sweet, no milk, no iced, no decaf
recommendation = coffee_expert_system(prefers_sweet=False, prefers_milk=False, prefers_strong=True, prefers_iced=False, prefers_decaf=False)
print(f"Sweet: False, Milk: False, Strong: True, Iced: False, Decaf: False -> Recommendation: {recommendation}")

# Test case 3: Iced, sweet, milk, no strong, no decaf
recommendation = coffee_expert_system(prefers_sweet=True, prefers_milk=True, prefers_strong=False, prefers_iced=True, prefers_decaf=False)
print(f"Sweet: True, Milk: True, Strong: False, Iced: True, Decaf: False -> Recommendation: {recommendation}")

# Test case 4: Decaf, no sweet, no milk, no strong, no iced
recommendation = coffee_expert_system(prefers_sweet=False, prefers_milk=False, prefers_strong=False, prefers_iced=False, prefers_decaf=True)
print(f"Sweet: False, Milk: False, Strong: False, Iced: False, Decaf: True -> Recommendation: {recommendation}")

# Test case 5: Default (no specific preferences)
recommendation = coffee_expert_system(prefers_sweet=False, prefers_milk=False, prefers_strong=False, prefers_iced=False, prefers_decaf=False)
print(f"Sweet: False, Milk: False, Strong: False, Iced: False, Decaf: False -> Recommendation: {recommendation}")

# Test case 6: Decaf, iced, milk, sweet
recommendation = coffee_expert_system(prefers_sweet=True, prefers_milk=True, prefers_strong=False, prefers_iced=True, prefers_decaf=True)
print(f"Sweet: True, Milk: True, Strong: False, Iced: True, Decaf: True -> Recommendation: {recommendation}")

# Test case 7: Strong, milk, sweet
recommendation = coffee_expert_system(prefers_sweet=True, prefers_milk=True, prefers_strong=True, prefers_iced=False, prefers_decaf=False)
print(f"Sweet: True, Milk: True, Strong: True, Iced: False, Decaf: False -> Recommendation: {recommendation}")
```
