```python
def factorial(n):
    """Calculates the factorial of a non-negative integer recursively."""
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")

number = 0
result = factorial(number)
print(f"The factorial of {number} is {result}")
```
